```python
import numpy as np
import pandas as pd
df = pd.read_csv("../data/raw/titanic3.csv")
df["money_wasted_cause_death"] = np.where(df["survived"]==0, df["fare"], 0) #np.where(condition, value_if_true, value_if_false)
df[["survived", "fare", "money_wasted_cause_death"]].head(10)
df["ticket/average money wasted"] = df["fare"]/(df["money_wasted_cause_death"].mean())
df[["fare", "money_wasted_cause_death", "ticket/average money wasted"]].head(10)

missing_count_fare = df["fare"].isnull().sum()
missing_count_age = df["age"].isnull().sum()
print(f'Missing values in fare column: {missing_count_fare}')
print(f'Missing values in age column: {missing_count_age}')

df["age"].fillna(df["age"].median(),inplace=True)
df["fare"].fillna(df["fare"].median(),inplace=True)
missing_count_fare = df["fare"].isnull().sum()                               # putting median value in missing areas
missing_count_age = df["age"].isnull().sum()                               # putting median value in missing areas
print(f'Missing values in fare column after imputation: {missing_count_fare}')
print(f'Missing values in age column after imputation: {missing_count_age}')     


df["fare_over_mean"] = df["fare"] / df["fare"].mean()
df["age_minus_median"] = df["age"] - df["age"].median()
df["fare_z"] = (df["fare"] - df["fare"].mean()) / df["fare"].std()
df[["fare", "fare_over_mean", "age", "age_minus_median", "fare_z"]].head(10)


df["is_child"]=np.where(df["age"]<18,1,0)
df[["age","is_child"]].head(10)
child_survived =np.sum(df["is_child"]&df["survived"])
percent_child_survived = child_survived/(np.sum(df["is_child"]))*100
adult_survived = np.sum((df["is_child"]==0)&df["survived"])
percent_adult_survived = adult_survived/(np.sum(df["is_child"]==0))*100
print(f'Percentage of children survived: {percent_child_survived}%')
print(f'Percentage of adults survived: {percent_adult_survived}%')

survival_by_child = df.groupby("is_child")["survived"].mean()
print(survival_by_child)
df.to_csv("../data/processed/titanic3_eda.csv", index=False)

```
